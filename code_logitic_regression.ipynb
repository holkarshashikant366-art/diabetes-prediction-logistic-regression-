# Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler

# Logistic Regression model
from sklearn.linear_model import LogisticRegression

# Evaluation metrics
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df=pd.read_csv("diabetes_dataset_v2.csv")
df.head()
```

# Check Dataset Info

```python
# Dataset shape (rows, columns)
df.shape

# Column names
df.columns
```

# Separate Features and Target

```python
x=df.drop("diabetes_risk",axis=1)
y=df["diabetes_risk"]
x.head()
```

# Train Test Split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Feature Scaling

```python
scaler=StandardScaler()
```

```python
x_train_scl=scaler.fit_transform(x_train)
x_test_scl=scaler.transform(x_test)
```

# Create and Train Model

```python
# Create Logistic Regression model object
model = LogisticRegression()

# Train model using training data
model.fit(x_train_scl, y_train)
```

# Make Predictions

```python
# Predict class labels for test data
y_pred = model.predict(x_test_scl)

y_pred[:10]
```

```python
# Calculate accuracy (Accuracy = correct predictions / Total predictions)
accuracy = accuracy_score(y_test, y_pred)

print("Model Accuracy:", accuracy)
```

# Confusion Matrix

```python
# Confusion Matrix shows prediction performance

cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)
```
